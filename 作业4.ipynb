```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] # SimHei

# 
data = pd.read_csv('.csv', encoding='GBK') # 

# 
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
data[''].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('')
plt.ylabel('')

plt.subplot(1, 2, 2)
data[''].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('')
plt.ylabel('')

plt.tight_layout()
plt.show()

# 
cross_table = pd.crosstab(data[''], data[''])
print(cross_table)

# 
correlation = data[''].astype('category').cat.codes.corr(data[''].astype('category').cat.codes)
print(f"{correlation}")

# 
if correlation > 0.5:explanation = ""
elif correlation < -0.5:explanation = ""
else:explanation = ""

print("", explanation)
```
