```python
import matplotlib.pyplot as plt

# 数据
x = [1,2,3,4,5]
y = [2,4,6,8,10]

# 创建线图
plt.plot(x,y)

# 添加标题和轴标签
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

#  显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt

#数据
categories = ['Categories1','Categories2','Categories3','Categories4','Categories5']
value = [10,15,7,10,5]

# 创建柱状图
plt.bar(categories,value)

# 添加标题和轴标签
plt.title('Simple Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Value')

# 显示图形
plt.show()
```
